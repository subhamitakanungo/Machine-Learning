```python
x=df.drop('MSRP',axis=1)
```

```python
x
```

```python
y=df['MSRP']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_train,y_train)*100
```

```python
model.score(x_test,y_test)*100
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy=r2_score(y_test,y_predict)*100
```

```python
accuracy
```

```python
plt.scatter(y_test,y_predict)
plt.show()
```

```python
path=model.cost_complexity_pruning_path(x_train, y_train)
```

```python
path
```

```python
alphas=path['ccp_alphas']
```

```python
alphas
```

```python
alphas.min()
```

```python
alphas.max()
```

```python
accuracy_train=[]
accuracy_test=[]

for i in alphas:
    model=DecisionTreeRegressor(ccp_alpha=i)
    model.fit(x_train,y_train)
    y_predict_train=model.predict(x_train)
    y_predict_test=model.predict(x_test)
    
    accuracy_train.append(r2_score(y_train,y_predict_train))
    accuracy_test.append(r2_score(y_test,y_predict_test))
```

```python
accuracy_train
```

```python
accuracy_test
```

```python
sns.lineplot(y=accuracy_train,x=alphas)
sns.lineplot(y=accuracy_test,x=alphas)
plt.show()
```
